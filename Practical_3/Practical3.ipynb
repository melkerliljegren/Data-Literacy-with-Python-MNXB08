# Challenge 1

How old is the newest house (woonhuis) with energy class B?

```python
(
    ut
    .filter(pl.col('house_type')=='woonhuis',
            pl.col('energy_label')=='B')
    .sort('build_year', descending= True)
    .select('build_year')
)
```

# Challenge 2

What are top-3 districts with the largest number of properties in energy class B or better? What is the number of missing values in the energy_label column of these district?

ğŸ’¡ To find out the number of houses in energy class B or better check out the meaning of columns energy_label and energy_eff. You might want to use both of them!

ğŸ’¡ The logical OR operator in Python is denoted by |

```python
(
    ut
    .filter((pl.col("energy_label").is_in(["A++", "A+", "A", "B"])) | (pl.col('energy_label').is_null() ))
    .group_by("district")
    .agg(n_properties=pl.len(),
         n_nulls=pl.col('energy_label').is_null().sum())
    .sort("n_properties", descending=True)
    .head()
)
```

```python
(
    ut
    .filter(pl.col("district").is_in(["Leidsche Rijn", "Zuid-West", "Vleuten-De Meern"]),
            pl.col("energy_label").is_null())
    .group_by("district")
    .agg(n_nulls=pl.len())
)
```

# Challenge 3

Create a great tables summary showing the proportion of energy-efficient homes (those with an energy label of A or better) for each decade from the 20th and 21st century.

ğŸ’¡ You can group houses by decade by flooring the build year to the start of the decade using integer division //

ğŸ’¡ The mean() of a binary variable is showing you the proportion of the â€œpositiveâ€ cases.

```python
(
    ut
    .filter((pl.col('build_year') >= 1900))
    .with_columns(((pl.col('build_year')//10)*10).alias('Decade'))
    .group_by('Decade')
    .agg(proportion = (pl.col('energy_eff').mean()))
    .sort(pl.col('Decade'), descending= True)
    .pipe(GT)
   

)
```

# Challenge 4

Reproduce this great table showing the minimum and maximum retail value per square meter by city and district, as well as the number of houses in each district. Table is arranged by the price range from lowest to highest. Which district has the most expensive housing?

ğŸ’¡ You can count number of records per group with pl.len()

```python
(
    ut
    .group_by("city", "district")
    .agg(n=pl.len(), 
         minimum=(pl.col("retail_value")/pl.col("house_area")).min(), 
         maximum=(pl.col("retail_value")/pl.col("house_area")).max())
    .with_columns(abs(pl.col("minimum")-pl.col("maximum")).alias("range"))
    .sort("range")
    .pipe(GT)
    .fmt_number(columns=cs.ends_with("imum"), decimals=2)
    .fmt_number(columns=pl.col("range"), decimals=2)
)
```

# Challenge 6

Create a great table summary showing top-3 most expensive houses in every district (in terms of asking price per sqm). Show house type, lot and house area, number of rooms and bathrooms. Arrange the houses from most expensive to least expensive within each district.

ğŸ’¡ There are number of ways of showing top-3 houses. I suggest you use the .rank().over('district').

ğŸ’¡ For extra fun, you can show â€˜woonhuisâ€™ and â€˜appartementâ€™ as pl.lit('house-chimney') and pl.lit('building') icons, respectively.

```python
(
    ut
    .with_columns(price_per_sqm = ((pl.col('asking_price'))/(pl.col('house_area'))))
    .with_columns(rank_price=pl.col('price_per_sqm').rank(descending=True).over('district'))
    .filter(pl.col('rank_price')<=3)
    .sort('district', 'rank_price')
)
```
